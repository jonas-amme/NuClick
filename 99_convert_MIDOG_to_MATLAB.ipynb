```python
from scipy.io import savemat

image_ids = list(set([o['image_id'] for o in data['annotations']]))

# save centroid coords in matlab format
for image_id in image_ids:
    bboxes = np.array([o['bbox'] for o in data['annotations'] if o['image_id'] == image_id])
    centroids = (bboxes[:, 2:] + bboxes[:, :2]) / 2
    filename = str(image_id).zfill(3)
    print('id', image_id, 'Boxes:', centroids.shape[0])
    savemat(f'./annotations/{filename}.mat', {'centroids': centroids})
```

```python
def mask2bbox(mask):

    # instances are encoded as different colors
    obj_ids = np.unique(mask)
    # first id is the background, so remove it
    obj_ids = obj_ids[1:]

    # split the color-encoded mask into a set
    # of binary masks
    masks = mask == obj_ids[:, None, None]

    # get bounding box coordinates for each mask
    num_objs = len(obj_ids)
    boxes = []
    for i in range(num_objs):
        pos = np.where(masks[i])
        xmin = np.min(pos[1])
        xmax = np.max(pos[1])
        ymin = np.min(pos[0])
        ymax = np.max(pos[0])
        boxes.append([xmin, ymin, xmax, ymax])

    return boxes
```

```python
import matplotlib.pyplot as plt
import os 
from PIL import Image
from tqdm.notebook import tqdm, trange


imgs_path = r'C:\Users\ammeling\Documents\projects\NuClick\images'
masks_path = r'C:\Users\ammeling\Documents\projects\NuClick\preds'
figure_path = r'C:\Users\ammeling\Documents\projects\NuClick\figures'

imgs = sorted(os.listdir(imgs_path))
masks = sorted(os.listdir(masks_path))

assert len(imgs) == len(masks)

PAD = 50

for _img, _mask in tqdm(zip(imgs, masks), total=len(imgs)):

    img = np.array(Image.open(os.path.join(imgs_path, _img)).convert('RGB'))
    mask = np.array(Image.open(os.path.join(masks_path, _mask)))

    boxes = mask2bbox(mask)

    if len(boxes) == 0:
        continue

    for box_idx, box in enumerate(boxes):

        with plt.ioff():

            xmin, ymin, xmax, ymax = box
            xmin, ymin, xmax, ymax = xmin-PAD, ymin-PAD, xmax+2*PAD, ymax+2*PAD

            if xmin < 0:
                xmin = 0
            elif xmax > img.shape[1]:
                xmax = img.shape[1]
            
            if ymin < 0:
                ymin = 0
            elif ymax > img.shape[0]:
                ymax = img.shape[0]

            fig, ax = plt.subplots(1, 2)
            ax[0].imshow(img[ymin:ymax, xmin:xmax])
            ax[1].imshow(mask[ymin:ymax, xmin:xmax])
            plt.suptitle(f'{_img}, mask {box_idx}')
            plt.savefig(os.path.join(figure_path, _img[:-5] + '_' + str(box_idx).zfill(3) + '.png'), bbox_inches='tight')
            plt.close(fig)       
```
